```python
import PCF8591 as ADC
import RPi.GPIO as GPIO
import time
import math

glodon_DO = 17 # 温度传感器Do管脚
GPIO.setmode(GPIO.BCM) # 管脚映射，采用BCM编码
GPIO.setwarnings(False)

# 初始化设置
def glodon_setup():
    ADC.setup(0x48)          # 设置PCF8591模块地址
    GPIO.setup(glodon_DO, GPIO.IN) # 温度传感器DO端口设置为输入模式

# 打印出温度传感器的提示信息
def glodon_Print(x):
    if x == 1:     # 正合适
        print ('')
        print ('***********')
        print ('* Just right ，Better~ *')
        print ('***********')
        print ('')
    if x == 0:    # 太热了
        print ('')
        print ('************')
        print ('* It is Too Hot! *')
        print ('************')
        print ('')

# 循环函数
def glodon_loop():
    glodon_status = 1   # 状态值
    glodon_tmp = 1      # 当前值
    while True:
        glodon_analogVal = ADC.read(0) # 读取AIN0上的模拟值
        glodon_Vr = 5 * float(glodon_analogVal) / 255 # 转换到5V范围
        glodon_Rt = 10000 * glodon_Vr / (5 - glodon_Vr)
        glodon_temp = 1/(((math.log(glodon_Rt / 10000)) / 3950) + (1 / (273.15+25)))
        glodon_temp = glodon_temp - 273.15
        print ('Current Temperature = ', glodon_temp, 'C')

        glodon_tmp = GPIO.input(glodon_DO) # 读取温度传感器数字端口

        if glodon_tmp != glodon_status: # 判断状态值发生改变
            glodon_Print(glodon_tmp)    # 打印出温度传感器的提示信息
            glodon_status = glodon_tmp  # 把当前状态值设置为比较状态值，避免重复打印； 

        time.sleep(0.2)                     # 延时 200ms

# 程序入口
if __name__ == '__main__':
    try:
        glodon_setup()  #初始化程序
        glodon_loop()   #循环函数
    except KeyboardInterrupt:  
        pass
```
